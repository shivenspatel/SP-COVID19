```python
import pandas as pd
import numpy as np
import folium
```

```python
#Pulling Data
cdata=pd.read_csv("https://raw.githubusercontent.com/nytimes/covid-19-data/master/live/us-counties.csv")
hdata=pd.read_csv("https://raw.githubusercontent.com/nytimes/covid-19-data/master/us-counties.csv")
cdata[cdata["county"] == "DuPage"]

county_data=pd.read_csv('c_03mr20.csv')

geojsonmap='us-counties.json'
```

```python
county_data
```

```python
cdata = cdata.fillna(0)
cdata['fips_str'] = cdata['fips'].astype(int).astype(str)
cdata
```

```python
center_data=pd.read_csv('table-1.csv')
center_data.set_index('state', inplace=True)
latitude = center_data['latitude']
longitude = center_data['longitude']
```

```python
statedata=cdata[cdata['state']=='Illinois']
statedata.set_index('county', inplace=True)
statedata.drop('Cook')
```

```python
m = folium.Map(location=[float(latitude.loc["IL"]), float(longitude.loc["IL"])], zoom_start=6)
```

```python
statecountydata=county_data[county_data['STATE'] == 'IL']
statecountydata.set_index('COUNTYNAME', inplace=True)
type(statecountydata.loc['DuPage']['LON'])
statecountydata
```

```python
folium.Choropleth(
    geo_data=geojsonmap,
    name='Positive Cases',
    data=statedata,
    columns=['fips_str', 'cases'],
    key_on='feature.fips',
    fill_color='OrRd',
    fill_opacity=0.5,
    line_opacity=0.2,
    legend_name='Positive COVID-19 Cases'
).add_to(m)
```

```python
for cn in statecountydata.index:
    folium.CircleMarker(location=(float(statecountydata.loc[cn]['LAT']), float(statecountydata.loc[cn]['LON'])), 
                        tooltip=cn, fill=True, fill_color='white', color='white').add_to(m)

m.save("ilcountymap.html")
```
